### JSON ëª¨ë“œ

```python
# ëª¨ë¸ ì¤€ë¹„
model = genai.GenerativeModel(
    "gemini-1.5-flash",
    generation_config={"response_mime_type": "application/json"}
)
```

```python
# í”„ë¡¬í”„íŠ¸ë¡œ JSON ì¶œë ¥ì„ ì§€ì‹œ
prompt = """ë‹¤ìŒ JSON ìŠ¤í‚¤ë§ˆë¥¼ ì‚¬ìš©í•´ì„œ ìœ ëª…í•œ ì¿ í‚¤ ë ˆì‹œí”¼ë¥¼ í•œêµ­ì–´ë¡œ 5ê°œë¥¼ ë¦¬ìŠ¤íŠ¸ì—… í•´ì£¼ì„¸ìš”.

Recipe = {'recipe_name': str}
Return: list[Recipe]"""

# ì¶”ë¡  ì‹¤í–‰
raw_response = model.generate_content(prompt)
print(type(raw_response.text))
print(raw_response.text)
```

### ì œì–´í˜• ë””ì½”ë”©ì„ ì‚¬ìš©í•œ JSON ì¶œë ¥

```python
import typing_extensions as typing

# ìŠ¤í‚¤ë§ˆ ê°ì²´ ì •ì˜
class Recipe(typing.TypedDict):
    recipe_name: str
```

```python
# ëª¨ë¸ ì¤€ë¹„
model = genai.GenerativeModel(
    model_name="models/gemini-1.5-pro"
)
```

```python
# ì œì–´í˜• ë””ì½”ë”©ì„ í†µí•œ JSON ì¶œë ¥ì„ ì§€ì‹œ
result = model.generate_content(
    "ìœ ëª…í•œ ì¿ í‚¤ ë ˆì‹œí”¼ë¥¼ í•œêµ­ì–´ë¡œ 5ê°œë¥¼ ë¦¬ìŠ¤íŠ¸ì—… í•´ì£¼ì„¸ìš”",
    generation_config=genai.GenerationConfig(
        response_mime_type="application/json",
        response_schema = list[Recipe]
    ),
    request_options={"timeout": 600},
)
print(type(result.text))
print(result.text)
```
