### ì±—ë´‡

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

# ëŒ€í™” ì´ë ¥ì„ ì €ì¥í•  ê³µê°„ì„ ì¤€ë¹„
store = {}

# ì„¸ì…˜ë§ˆë‹¤ ëŒ€í™” ì´ë ¥ì„ ê°€ì ¸ì˜¤ê¸°
def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

# ì±—ë´‡ ì •ì˜
chatbot = RunnableWithMessageHistory(llm, get_session_history)
```

```python
# ì„¤ì • ì¤€ë¹„
config = {"configurable": {"session_id": "abc1"}}
```

```python
# ëŒ€í™”
response = chatbot.invoke(
    [("human", "ì•ˆë…•í•˜ì„¸ìš”! ì œ ì´ë¦„ì€ ìŠ¹ë¯¼ì…ë‹ˆë‹¤.")],
    config=config,
)
response.content
```

```python
# ê³¼ê±° ëŒ€í™” ë‚´ìš©ì— ê´€í•´ ì§ˆë¬¸
response = chatbot.invoke(
    [("human", "ë‚´ ì´ë¦„ì€?")],
    config=config,
)
response.content
```

### ì»¤ìŠ¤í…€ ì§€ì‹œ

```python
from langchain_core.prompts import ChatPromptTemplate

# í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ ì¤€ë¹„
prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "ë‹¹ì‹ ì€ í›Œë¥­í•œ AIë¹„ì„œì…ë‹ˆë‹¤. ë§ ëì— '~ëë‹ˆë‹¤.'ë¥¼ ë¶™ì—¬ ëŒ€ë‹µí•´ì£¼ì„¸ìš”."),
        ("placeholder", "{messages}"),
    ]
)

# ì²´ì¸ ì¤€ë¹„
chain = prompt_template | llm
```

```python
# ì±—ë´‡ ì¤€ë¹„
chatbot = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages",
)
```

```python
# ì„¤ì • ì¤€ë¹„
config = {"configurable": {"session_id": "abc2"}}
```

```python
# ëŒ€í™”
response = chatbot.invoke(
    {"messages": [("human", "ì•ˆë…•í•˜ì„¸ìš”! ì œ ì´ë¦„ì€ ìŠ¹ë¯¼ì…ë‹ˆë‹¤.")]},
    config=config
)
response.content
```

### ëŒ€í™” ì´ë ¥ ê´€ë¦¬

```python
from langchain_core.runnables import RunnablePassthrough

# ë©”ì‹œì§€ í•„í„° ì¤€ë¹„
def filter_messages(messages, k=4):
    return messages[-k:]

# ì²´ì¸ ì¤€ë¹„
chain = (
    RunnablePassthrough.assign(messages=lambda x: filter_messages(x["messages"]))
    | prompt_template
    | llm
)
```

```python
# ì±—ë´‡ ì¤€ë¹„
chatbot = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages",
)
```

```python
# ì„¤ì • ì¤€ë¹„
config = {"configurable": {"session_id": "abc3"}}
```

```python
# ëŒ€í™”
response = chatbot.invoke(
    {"messages": [("human", "ì•ˆë…•í•˜ì„¸ìš”! ì œ ì´ë¦„ì€ ìŠ¹ë¯¼ì…ë‹ˆë‹¤.")]},
    config=config,
)
response.content
```
