### Faiss 사용 순서

```python
# 패키지 설치
!pip install llama-index-vector-stores-faiss
!pip install faiss-cpu
```

```python
from llama_index.core import SimpleDirectoryReader

# 문서 불러오기(data 폴더에 문서를 먼저 업로드)
documents = SimpleDirectoryReader("data").load_data()
```

```python
import faiss

# Faiss 인덱스 준비
faiss_index = faiss.IndexFlatL2(1024)
```

```python
from llama_index.core import (VectorStoreIndex, StorageContext)
from llama_index.vector_stores.faiss import FaissVectorStore

# 인덱스 준비
vector_store = FaissVectorStore(faiss_index=faiss_index)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context
)
```

```python
# 쿼리 엔진 준비
query_engine = index.as_query_engine()
```

```python
# 질의 응답
response = query_engine.query("은비의 나이는?")
print(response)
```

### Pinecone을 사용한 질의 응답

```python
# 패키지 설치
!pip install llama-index-vector-stores-pinecone
!pip install pinecone-client
```

```python
from llama_index.core import SimpleDirectoryReader

# 문서 불러오기(data 폴더에 문서를 먼저 업로드)
documents = SimpleDirectoryReader("data").load_data()
```

```python
from google.colab import userdata
import os

# 환경 변수 준비(좌측 하단의 열쇠 아이콘으로 PINECONE_API_KEY 설정)
os.environ["PINECONE_API_KEY"] = userdata.get("PINECONE_API_KEY")
```

```python
from pinecone import Pinecone

# Pinecone 객체 준비
pc = Pinecone(
    api_key=os.environ["PINECONE_API_KEY"]
)
```

```python
# 기존 Pinecone 인덱스 가져오기
indexes = pc.list_indexes()
print(indexes)
```

```python
from pinecone import PodSpec

# Pinecone 인덱스 생성
if "quickstart-index" not in indexes:
    pc.create_index(
        "quickstart-index",
        dimension=1024,
        metric="dotproduct",
        spec=PodSpec(
            environment="us-west1-gcp",
            pod_type="p1.x1",
            pods=1
        )
    )
```

```python
# Pinecone 인덱스 준비
pc_index = pc.Index("quickstart-index")
```

```python
from llama_index.vector_stores.pinecone import PineconeVectorStore
from llama_index.core import VectorStoreIndex

# 인덱스 준비
vector_store = PineconeVectorStore(
    pinecone_index=pc_index,
    text_key="content"
)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context
)
```

```python
# 쿼리 엔진 준비
query_engine = index.as_query_engine()
```

```python
# 질의 응답
response = query_engine.query("은비의 나이는?")
print(response)
```
